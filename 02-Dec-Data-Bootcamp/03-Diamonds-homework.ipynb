```python
import pandas as pd
import numpy as np
```

```python
#using functions
filename = "datasets/diamonds.csv"
df = pd.read_csv(filename)
df.head()
```

```python
def multiplay(x,y,z):
    return x*y*z
df['product'] = multiplay(df['x'],df['y'],df['z'])
df.head()
```

```python
name = ['eehee', 'deep', 'shallow']
bins = [0,55,70,100]
df['newcolumn'] = pd.cut(df['depth'],bins,labels=name)
df.sort_values('depth', ascending = False)
```

```python
def score(i):
    if i == 'Ideal':
        return 'A'
    if i == 'Premium':
        return 'B'
    if i == 'Good':
        return 'C'
    if i == 'Very Good':
        return 'D'
    if i == 'Fair':
        return 'E'
df['cut-val'] = df['cut'].apply(score)
df.tail()
```
